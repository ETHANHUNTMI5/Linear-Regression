## Surface plot and Data Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a= np.array([1,2,3])
#b= np.array([4,5,6,7])

a = np.arange(-1,1,0.02)
b=a
```

```python
a,b = np.meshgrid(a,b)
#print(a)
#print(b)
```

```python
fig = plt.figure()
axes= fig.gca(projection = "3d")
axes.plot_surface(a,b,(a**2-b**2),cmap='rainbow')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
## Contour
fig = plt.figure()
axes = fig.gca(projection="3d")
axes.contour(a,b,a**2+b**2,cmap="rainbow")
plt.xlabel("X")
plt.title("CONTOUR PLOT")
plt.show()
```
