## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y =(X-5)**2
print(X)
print(Y)
```

## Given a function f , find value of x that minimises f

```python
##Visualisation
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)")
plt.show()
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)")
x=0
lr=0.1
error=[]

for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    print(x)
    plt.scatter(x,y)
    
```
